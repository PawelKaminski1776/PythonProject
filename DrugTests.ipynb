```python
import pandas as pd

dp = pd.read_csv('drugs.csv')

unique_years = dp['Year'].unique()
print(unique_years)
```

```python
Total_Cocaine_Abuse = dp['Totals.Illicit Drugs.Cocaine Used Past Year.12-17'].sum() + dp['Totals.Illicit Drugs.Cocaine Used Past Year.18-25'].sum() + dp['Totals.Illicit Drugs.Cocaine Used Past Year.26+'].sum()

print(Total_Cocaine_Abuse)
```

```python
target_year = 2016

yearly_data = dp[dp['Year'] == target_year]

total_cocaine_abuse_yearly = yearly_data['Totals.Illicit Drugs.Cocaine Used Past Year.12-17'].sum() + yearly_data['Totals.Illicit Drugs.Cocaine Used Past Year.18-25'].sum() + yearly_data['Totals.Illicit Drugs.Cocaine Used Past Year.26+'].sum()

print(f'Total Cocaine Abuse in {target_year}: {total_cocaine_abuse_yearly}')
```

```python
import pandas as pd

years = dp['Year'].unique()

for year in years:
    yearly_data = dp[dp['Year'] == year]
    total_cocaine_abuse_yearly = yearly_data[['Totals.Illicit Drugs.Cocaine Used Past Year.12-17', 'Totals.Illicit Drugs.Cocaine Used Past Year.18-25', 'Totals.Illicit Drugs.Cocaine Used Past Year.26+']].sum().sum()
    print(f'Total Cocaine Abuse in {year}: {total_cocaine_abuse_yearly}')
```

```python
# Total Cocaine Abuse over the year plot

import pandas as pd
import matplotlib.pyplot as plt

years = dp['Year'].unique()
total_abuse_per_year = []

for year in years:
    yearly_data = dp[dp['Year'] == year]
    total_cocaine_abuse_yearly = yearly_data[['Totals.Illicit Drugs.Cocaine Used Past Year.12-17', 'Totals.Illicit Drugs.Cocaine Used Past Year.18-25', 'Totals.Illicit Drugs.Cocaine Used Past Year.26+']].sum().sum()
    total_abuse_per_year.append(total_cocaine_abuse_yearly)

# Plotting
plt.plot(years, total_abuse_per_year)
plt.title('Total Cocaine Abuse Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Cocaine Abuse')
plt.show()
```

```python
# Total Marijuana Abuse over the year plot

import pandas as pd
import matplotlib.pyplot as plt

years = dp['Year'].unique()
total_abuse_per_year = []

for year in years:
    yearly_data = dp[dp['Year'] == year]
    total_Marijuana_abuse_yearly = yearly_data[['Totals.Marijuana.Used Past Year.12-17', 'Totals.Marijuana.Used Past Year.18-25', 'Totals.Marijuana.Used Past Year.26+']].sum().sum()
    total_abuse_per_year.append(total_Marijuana_abuse_yearly)

# Plotting
plt.plot(years, total_abuse_per_year)
plt.title('Total Marijuana Abuse Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Marijuana Abuse')
plt.show()
```
